## **Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Style setting**

```python
sns.set(style="whitegrid")
```

# **Data Loading**

```python
df = pd.read_csv("menu.csv")
print("Shape of dataset:", df.shape)
df.head()
```

# **Data Overview**

**1.Basic info.**

```python
print("\n--- Basic Info ---")
print(df.info())
```

**2.Statistical summary**

```python
print("\n--- Statistical Summary ---")
print(df.describe())
```

# **Analysis of Nutrient**

**Distribution of calories**

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Calories"], bins=30, color="blue")
plt.title("Calories Distribution")
plt.xlabel("Calories")
plt.ylabel("Count")
plt.show()
```

**Highest Calorie Items**

```python
top10_calories = df.nlargest(10, "Calories")[['Item','Category','Calories']]
print("\nTop 10 Highest Calorie Items:\n", top10_calories)
```

**Highest Protein items**

```python
top10_protein = df.nlargest(10, "Protein")[['Item','Category','Protein']]
print("\nTop 10 Highest Protein Items:\n", top10_protein)
```

**Distribution by Category**

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Category", y="Calories", data=df)
plt.title("Calorie Distribution by Category")
plt.xticks(rotation=45)
plt.show()
```

**Nutritional values**

```python
nutrient_cols = df.select_dtypes(include=['int64','float64']).columns
corr_matrix = df[nutrient_cols].corr()


plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Nutritional Values")
plt.show()
```

**Conclusion**

```python
print("\nInsights:")
print("1. Categories like Beef & Pork tend to have the highest average calories.")
print("2. Chicken McNuggets (40 piece) is the top protein-heavy item (87g protein).")
print("3. Correlation shows Calories strongly linked with Fat & Carbs.")
```
